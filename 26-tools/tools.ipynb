## Custom tool

```python
from langchain_core.tools import tool

# step 1 - define function

def multiply(a,b):
    """Multiply two numbers"""       ## doc string
    return a*b
```

```python
# Step 2 - add type hints
def multiply(a:int ,b:int) -> int :
    """Multiply two numbers"""    
    return a*b
```

```python
# Step 3 - add tool decorator

@tool
def multiply(a:int ,b:int) -> int :
    """Multiply two numbers"""    
    return a*b
```

```python
result = multiply.invoke({"a":2,"b":3})
print(result)
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
# Method 2 - using structured Tool
```

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field

class MultiplyInput(BaseModel):
    a: int = Field(description="First number to add", required = True)
    b: int = Field(description="Second number to add", required = True)


def multiply_func(a:int, b:int) ->int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func= multiply_func,
    name = "Multiply",
    description= "Multiply two numbers",
    args_schema= MultiplyInput
)
```

```python
result = multiply_tool.invoke({"a":2,"b":3})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
# using base tool class
```

```python
from langchain_core.tools import BaseTool
from typing import Type

# arg schema using pydantic
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
class MultiplyTool(BaseTool):
    name:str =  "multiply"
    description: str = "multiply two number"
    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a:int, b:int)->int :
        return a*b
```

```python
mul_tool = MultiplyTool()
```

```python
result = mul_tool.invoke({"a":2,"b":3})
print(result)
print(mul_tool.name)
print(mul_tool.description)
```

## Toolkits

```python
# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolKit:
    def get_tool(self):
        return [add, multiply]
```

```python
toolkit = MathToolKit()
tools = toolkit.get_tool()

for tool in tools:
    print(tool.name," -> ",tool.description)
```
