```python
# types of loader
# 1) text loader
# 2) pdf loader
# 3) web base loader
# 4) arxiv loader
# 5) wikipedia loader
```

```python
## text loader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("../speech.txt")
loader
```

```python
text = loader.load()
text
```

```python
## pdf loader

from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("../paper 1.pdf")

doc = loader.load()
doc
```

```python
type(doc[0])
```

```python
## web base

from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_paths = ("https://arxiv.org/abs/1706.03762",),
                       bs_kwargs = dict(parse_only = bs4.SoupStrainer(
                           class_=("title mathjax","abstract mathjax")
                       )))
```

```python
loader.load()
```

```python
## arxiv

from langchain_community.document_loaders import ArxivLoader

loader = ArxivLoader(query = "1706.03762", load_max_docs = 2)
docs = loader.load()
docs
```

```python
len(docs)
```

```python
## wikipedia loader

from langchain_community.document_loaders import WikipediaLoader

docs = WikipediaLoader(query = "Bangladesh", load_max_docs = 2).load()

print(len(docs))
print(docs)
```
