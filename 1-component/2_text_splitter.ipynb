#### Text Splitting from Documents- RecursiveCharacter Text Splitters


```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("../paper 1.pdf")
docs = loader.load()
docs
```

```python
type(docs[0])
```

##### How to recursively split text by characters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)

final_document = text_splitter.split_documents(docs)
final_document
```

```python
print(final_document[0])
print(final_document[1])
```

```python
## text loader

from langchain_community.document_loaders import TextLoader

docs = TextLoader('../speech.txt').load()
docs
```

```python
speech = ""
with open('../speech.txt') as file:
    speech = file.read()

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 30)

text = text_splitter.create_documents([speech])
```

```python
print(text[0])
print(text[1])
```

## character text splitter

```python
## text loader

from langchain_community.document_loaders import TextLoader

docs = TextLoader('../speech.txt').load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size = 100, chunk_overlap = 20)
text_splitter.split_documents(docs)
```

```python
speech = ""

with open("../speech.txt") as file:
    speech = file.read()

text_splitter = CharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
text = text_splitter.create_documents([speech])
print(text[0])
print(text[1])
```
